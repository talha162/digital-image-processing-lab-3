```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage
from skimage.color import rgb2gray
```

```python
image=cv2.imread('/content/flower.jpeg')
```

```python
plt.imshow(image)
```

```python
levels=256
```

```python
newimage=levels-image[:,:,:]
```

```python
plt.imshow(newimage)
```

```python
histogram0=cv2.calcHist([image],[0],None,[256],[0,256])
plt.plot(histogram0)
plt.show()
```

[before inverting histogram]

```python
histogram0=cv2.calcHist([image],[0],None,[256],[0,256])
plt.plot(histogram0)
plt.show()
```

```python
plt.hist(image.ravel(), bins=256,) #calculating histogram
plt.show()
```

[after inverting histogram]

```python
plt.hist(newimage.ravel(), bins=256,) #calculating histogram
plt.show()
```

```python
histogram0=cv2.calcHist([newimage],[0],None,[256],[0,256])
plt.plot(histogram0)
plt.show()
```

```python
sndimage=cv2.imread('/content/image2.jpg')
sndimage=rgb2gray(sndimage)
```

```python
gamma=.1
gamma_corrected=np.array(255*(sndimage/255)**gamma,dtype='uint8')
```

```python
histogram1=cv2.calcHist([gamma_corrected],[0],None,[256],[0,256])
plt.plot(histogram1)
plt.show()
```

```python
plt.imshow(sndimage)
```

```python
plt.imshow(gamma_corrected)
```
